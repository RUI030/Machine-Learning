```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
M = [5, 10, 15, 20, 25, 30]
```

# 1. Fitting Curve
* The fitting curve of the $3^{rd}$ feature for M = \[5, 10, 15, 25, 30\] 

```python
title = '1. Fitting Curve'
path = "homework/1_M"
spath = "img/" + title + ".png"
plt.figure(figsize=(6, 4), dpi=300)
x = np.linspace(-3, 3, 1000)
fit = []

for m in M:
    name = path + str(m) + ".csv"
    _fit = np.loadtxt(name, delimiter=',')
    fit.append(_fit)
    plt.plot(x, _fit, label=f'M = {m}')  # Add a label for each line

# Add labels, legends, and show the plot outside of the loop
plt.title(title)
plt.xlabel('input')
plt.ylabel('y')
plt.legend()
plt.savefig(spath)
plt.show()
```

# 2. Evaluation
### a. Mean Square Error

```python
title = '2-1. Mean Square Error'
spath = "img/" + title + ".png"
name = "homework/2_1.csv"
plt.figure(figsize=(6, 4), dpi=300)
x = np.array(M)
mse = np.loadtxt(name, delimiter=',')
train = []
valid = []
for i in range(12):
    if(i%2):
        valid.append(mse[i])
    else:
        train.append(mse[i])

plt.plot(x, train, label = 'train')  
plt.plot(x, valid, label = 'valid')  
plt.title(title)
plt.xlabel('M')
plt.ylabel('MSE')
plt.legend()
plt.savefig(spath)
plt.show()
```

### b. Accuracy

```python
title = '2-2. Accuracy'
spath = "img/" + title + ".png"
name = "homework/2_2.csv"
plt.figure(figsize=(6, 4), dpi=300)
x = np.array(M)
mse = np.loadtxt(name, delimiter=',')
train = []
valid = []
for i in range(12):
    if(i%2):
        valid.append(mse[i]*100)
    else:
        train.append(mse[i]*100)

plt.plot(x, train, label = 'train')  
plt.plot(x, valid, label = 'valid')  
plt.title(title)
plt.xlabel('M')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.savefig(spath)
plt.show()
```
